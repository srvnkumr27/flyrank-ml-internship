## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
# ## My lane as an ML task
#
# Task Type: Classification
#
# I chose a classification task because the objective is to predict whether a content item is likely to be high-performing or not. The output belongs to one of two classes (high-performing or low-performing), making classification the most appropriate ML approach.
#
# The model will learn patterns from historical data and use them to support future content decisions.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
## Success Metric

#The primary evaluation metric is **F1 Score**.

#Content performance datasets are often imbalanced because relatively few items become highly successful. F1 Score balances Precision and Recall, making it a better indicator of model quality than Accuracy alone.

#A good model would achieve an F1 Score above 0.80 while maintaining balanced Precision and Recall.
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd

# Example dataset
df = pd.DataFrame({
    "content_id":[101,102,103,104,105],
    "title":[
        "Top AI Tools",
        "SEO Guide",
        "Python Basics",
        "Machine Learning Tutorial",
        "Data Science Tips"
    ],
    "topic":[
        "AI",
        "SEO",
        "Programming",
        "Machine Learning",
        "Data Science"
    ],
    "word_count":[950,1700,700,1500,1200],
    "ctr":[8.2,2.8,5.6,9.1,4.4]
})

df["high_performance"] = (df["ctr"] >= 5).astype(int)

df
```

## Unit of Analysis

One row represents one published content item.

Each row contains measurable features describing a single piece of content, including topic, title, word count, and click-through rate. These features are used to predict whether the content will perform well.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# ## Why ML Beats a Fixed Rule
#
# Content performance depends on many interacting factors, including topic, title quality, publishing time, content length, keywords, and audience behavior.
#
# A simple rule such as:
#
# "If word count is greater than 1000, then predict success"
#
# cannot capture these complex relationships.
#
# Machine learning learns patterns from historical observations and combines multiple features to make more accurate predictions. This makes the system adaptable as user behavior and content trends change over time.
#
# The prediction supports decision-making by helping identify which content is more likely to perform well before publication or promotion.
```
